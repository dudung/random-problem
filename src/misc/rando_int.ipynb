# random_int
random integer for miscellaneous examples

```python
import random as rnd

x = []
xmin = 10
xmax = 100
for i in range(100):
    x.append(rnd.randint(xmin, xmax))
print(x)
```

```python
import matplotlib.pyplot as plt

plt.hist(x, bins=10)
plt.xticks([*range(10, 101, 10)])
plt.grid()
plt.show()
```
