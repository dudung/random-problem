# random_int
usage of `randint()` and its features

```python
import random as rnd

x = []
xmin = 10  # inclusive
xmax = 100 # exclusive
for i in range(100):
    x.append(rnd.randint(xmin, xmax))
print(x)
```

```python
import matplotlib.pyplot as plt

ticks = [*range(10, 101, 10)]
print(ticks)
plt.hist(x, bins=9, range=(10,100))
plt.xticks(ticks)
plt.grid()
plt.xlabel('x')
plt.ylabel('N')
plt.show()
```

```python
import random as rnd
import matplotlib.pyplot as plt

xmin = 10
xmax = 99
N  = 100
width = 20
xticks = [*range(xmin, xmax + 2, width)]
yticks = [0, 5, 10, 15, 20]

rows = 3
cols = 4
i = 0
plt.figure(figsize=(10,6))
for r in range(1, rows+1):
    for c in range(1, cols+1):
        x = []
        for n in range(N):
            x.append(rnd.randint(xmin, xmax + 1))
        i += 1
        plt.subplot(rows, cols, i)
        plt.hist(x, bins=9, range=(10,100))
        plt.xticks(xticks)
        plt.yticks(yticks)
        plt.grid()
plt.show()
```

```python
import random as rnd
import matplotlib.pyplot as plt

xmin = 10
xmax = 99
N  = 100
width = 20
xticks = [*range(xmin, xmax + 2, width)]
yticks = [0, 5, 10, 15, 20]

rows = 3
cols = 4
i = 0
plt.figure(figsize=(10,6))
for r in range(1, rows+1):
    for c in range(1, cols+1):
        x = []
        for n in range(N):
            rnd.seed(n) # it uses always the same seeds order
            x.append(rnd.randint(xmin, xmax + 1))
        i += 1
        plt.subplot(rows, cols, i)
        plt.hist(x, bins=9, range=(10,100))
        plt.xticks(xticks)
        plt.yticks(yticks)
        plt.grid()
plt.show()
```
